```python
# Write a program that prints the numbers from 1 to 100. 
# But for multiples of three print “Trademark” instead of the number and for the multiples of five print “Vision”. 
# For numbers which are multiples of both three and five print “TrademarkVision". has context menu
for i in range(1,101):
    if i % 3 == 0 and i % 5 == 0:
        print('TrademarkVision')
    elif i % 3 == 0:
        print('Trademark')
    elif i % 5 == 0:
        print('Vision')
    else:
        print(i)
    # print(i)
```
